## Total Drafts and Users

```python

# Step 1: Calculate cumulative number of drafts
collective_df['draft_date'] = pd.to_datetime(collective_df['draft_time']).dt.date
drafts_by_day = collective_df.groupby('draft_date')['draft_id'].nunique()
cumulative_drafts = drafts_by_day.cumsum()

# Step 2: Calculate percentage of cumulative distinct users
distinct_users_by_day = collective_df.groupby('draft_date')['user_id'].nunique()
cumulative_distinct_users = distinct_users_by_day.cumsum()
cumulative_distinct_users_percentage = (cumulative_distinct_users / cumulative_distinct_users.iloc[-1]) * 100

# Create a figure with a secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add the cumulative drafts trace
fig.add_trace(go.Scatter(x=cumulative_drafts.index, y=cumulative_drafts, mode='lines+markers', name='Total Drafts Over Time', marker=dict(color='RoyalBlue')), secondary_y=False)

# Add the cumulative distinct users percentage trace
fig.add_trace(go.Scatter(x=cumulative_distinct_users_percentage.index, y=cumulative_distinct_users_percentage, mode='lines+markers', name='Cumulative Distinct Users %', marker=dict(color='FireBrick')), secondary_y=True)

# Update layout for better visualization
fig.update_layout(title_text='BBM4: Drafts and Cumulative Distinct Users Over Time', title_x=0.5)

# Update x-axis title
fig.update_xaxes(title_text='Date')

# Update y-axes titles
fig.update_yaxes(title_text='Cumulative Number of Drafts', secondary_y=False)
fig.update_yaxes(title_text='Percentage of Cumulative Distinct Users (%)', secondary_y=True, tickformat='.2f')

fig.show()
```

```python

# Create a figure with a secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add the cumulative drafts trace
fig.add_trace(go.Scatter(x=cumulative_drafts.index, y=cumulative_drafts, mode='lines+markers', name='Total Drafts Over Time', marker=dict(color='RoyalBlue')), secondary_y=False)

# Add the cumulative distinct users percentage trace
fig.add_trace(go.Scatter(x=cumulative_distinct_users_percentage.index, y=cumulative_distinct_users_percentage, mode='lines+markers', name='Cumulative Distinct Users %', marker=dict(color='FireBrick')), secondary_y=True)

# Update layout for better visualization
fig.update_layout(
    title_text='BBM4: Drafts and Cumulative Distinct Users Over Time',
    title_x=0.5,
    legend=dict(x=1.1, y=1),  # Adjusting the legend position
)

# Update x-axis title
fig.update_xaxes(title_text='Date')

# Update y-axes titles
fig.update_yaxes(title_text='Cumulative Number of Drafts', secondary_y=False)
fig.update_yaxes(title_text='Percentage of Cumulative Distinct Users (%)', secondary_y=True, tickformat='.2f')

fig.show()

## Log Slope Calc

# Calculate the log of the series
log_cumulative_drafts = np.log(cumulative_drafts)
log_cumulative_distinct_users_percentage = np.log(cumulative_distinct_users_percentage)

# Calculate the slope (rate of change) for each log-transformed series
# Note: linregress expects numerical x values, so we convert dates to ordinal numbers
drafts_dates_ordinal = np.array([d.toordinal() for d in log_cumulative_drafts.index])
users_dates_ordinal = np.array([d.toordinal() for d in log_cumulative_distinct_users_percentage.index])

# Ensure no NaN values before regression
drafts_dates_ordinal_clean, log_cumulative_drafts_clean = drafts_dates_ordinal[~np.isnan(log_cumulative_drafts)], log_cumulative_drafts[~np.isnan(log_cumulative_drafts)]
users_dates_ordinal_clean, log_cumulative_distinct_users_percentage_clean = users_dates_ordinal[~np.isnan(log_cumulative_distinct_users_percentage)], log_cumulative_distinct_users_percentage[~np.isnan(log_cumulative_distinct_users_percentage)]

# Linear regression to find slope
slope_drafts, _, _, _, _ = linregress(drafts_dates_ordinal_clean, log_cumulative_drafts_clean)
slope_users, _, _, _, _ = linregress(users_dates_ordinal_clean, log_cumulative_distinct_users_percentage_clean)

# Print the slopes
print(f"Log slope of cumulative drafts: {slope_drafts}")
print(f"Log slope of cumulative distinct users percentage: {slope_users}")

```

## Draft Activity Overview

```python

# Group by 'draft_date' and calculate the number of distinct 'draft_id's and 'user_id's for each day
daily_drafts_and_users = collective_df.groupby('draft_date').agg(
    total_drafts=('draft_id', 'nunique'),
    total_users=('user_id', 'nunique')
)

# Calculate the average number of drafts per distinct user for each day
daily_drafts_and_users['avg_drafts_per_user'] = daily_drafts_and_users['total_drafts'] / daily_drafts_and_users['total_users']

# Create subplots
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.1, subplot_titles=('Total Number of Drafts Each Day', 'Average Number of Drafts per Distinct User Each Day'))

# Add trace for the total number of drafts each day
fig.add_trace(go.Scatter(
    x=daily_drafts_and_users.index,
    y=daily_drafts_and_users['total_drafts'],
    mode='lines+markers',
    name='Total Drafts'
), row=1, col=1)

# Add trace for the average number of drafts per distinct user each day
fig.add_trace(go.Scatter(
    x=daily_drafts_and_users.index,
    y=daily_drafts_and_users['avg_drafts_per_user'],
    mode='lines+markers',
    name='Avg Drafts per User'
), row=2, col=1)

# Update layout for better visualization
fig.update_layout(
    title_text='<b>BBM4: Draft Activity Overview<b>',
    title_x=0.5,
    #xaxis_title='Date',
    height=600,
    showlegend=False
)

# Update y-axes titles
fig.update_yaxes(title_text='Total Drafts', row=1, col=1)
fig.update_yaxes(title_text='Avg Drafts per User', row=2, col=1)

fig.show()


# Ensure the index is a proper datetime format and sort it
daily_drafts_and_users.index = pd.to_datetime(daily_drafts_and_users.index)
daily_drafts_and_users.sort_index(inplace=True)

# Log transform the series
log_total_drafts = np.log(daily_drafts_and_users['total_drafts'])
log_avg_drafts_per_user = np.log(daily_drafts_and_users['avg_drafts_per_user'])

# Convert dates to ordinal numbers for regression analysis
dates_ordinal = np.array([date.toordinal() for date in daily_drafts_and_users.index])

# Linear regression on log-transformed total drafts
slope_total, intercept_total, r_value, p_value, std_err = stats.linregress(dates_ordinal, log_total_drafts)

# Linear regression on log-transformed average drafts per user
slope_avg, intercept_avg, r_value_avg, p_value_avg, std_err_avg = stats.linregress(dates_ordinal, log_avg_drafts_per_user)

# Print the slopes (growth rates)
print(f"Log slope of total drafts: {slope_total}")
print(f"Log slope of avg drafts per user: {slope_avg}")

# Calculate and print the delta between the slopes
delta_slopes = slope_avg - slope_total
print(f"Delta between the slopes: {delta_slopes}")
```

## Total Drafts vs New Users

```python
# Convert 'draft_time' to datetime and extract the date part to a new 'draft_date' column
collective_df['draft_time'] = pd.to_datetime(collective_df['draft_time'])
collective_df['draft_date'] = collective_df['draft_time'].dt.date

# Sort by draft_date to ensure chronological order
collective_df.sort_values('draft_date', inplace=True)

# Tracking seen users and counting new users per day
seen_users = set()
new_users_per_day = pd.Series(dtype=int)

for date, group in collective_df.groupby('draft_date'):
    new_users = group['username'].apply(lambda x: x not in seen_users).sum()
    seen_users.update(group['username'])
    new_users_per_day.loc[date] = new_users

# Calculate overall users per day
overall_users_per_day = collective_df.groupby('draft_date')['draft_id'].nunique()

# Plotly graph to compare overall users vs new users
fig = go.Figure()

# Overall users
fig.add_trace(go.Scatter(x=overall_users_per_day.index, y=overall_users_per_day, mode='lines+markers', name='Total Drafts'))

# New users
fig.add_trace(go.Scatter(x=new_users_per_day.index, y=new_users_per_day, mode='lines+markers', name='New Users'))

# Update layout for better visualization
fig.update_layout(
    title='Total Drafts Overall vs. New Users',
    xaxis_title='Date',
    yaxis_title='Number of Users',
    xaxis=dict(tickmode='auto', nticks=20, tickformat='%Y-%m-%d'),
    yaxis=dict(range=[0, max(overall_users_per_day.max(), new_users_per_day.max()) + 1])
)

fig.show()
```

## Ratio of New Users to Total Daily Users Over Time

```python
import pandas as pd
import plotly.graph_objects as go

# Ensure 'draft_time' is in datetime format in collective_df
collective_df['draft_time'] = pd.to_datetime(collective_df['draft_time'])
collective_df['draft_date'] = collective_df['draft_time'].dt.date

# Sort by draft_date to ensure chronological order
collective_df.sort_values('draft_date', inplace=True)

# Initialize a set to track seen users and a series to count new users per day
seen_users = set()
new_users_per_day = pd.Series(0, index=pd.to_datetime(collective_df['draft_date']).dt.date.unique())

# Calculate new users per day
for date, group in collective_df.groupby('draft_date'):
    current_users = set(group['username'])
    new_users = current_users - seen_users
    new_users_per_day[date] = len(new_users)
    seen_users.update(current_users)

# Calculate total daily users
total_daily_users = collective_df.groupby('draft_date')['username'].nunique()

# Calculate the ratio of new users to total daily users
new_to_total_ratio = new_users_per_day / total_daily_users

# Plotly graph to compare total daily users vs new users
fig = go.Figure()

# Ratio of new users to total daily users
fig.add_trace(go.Scatter(x=new_to_total_ratio.index, y=new_to_total_ratio, mode='lines+markers', name='New to Total Users Ratio'))

# Update layout for better visualization
fig.update_layout(
    title='Ratio of New Users to Total Daily Users Over Time',
    xaxis_title='Date',
    yaxis_title='New to Total Users Ratio',
    xaxis=dict(tickmode='auto', nticks=20, tickformat='%Y-%m-%d'),
    yaxis=dict(tickformat='.2%')  # Format as percentage
)

fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_average_drafts_per_user_vs_fill_rate(collective_df):
    # Calculate the total number of drafts per day
    total_drafts_per_day = collective_df.groupby('draft_date')['draft_entry_id'].nunique()

    # Calculate the total number of drafts
    total_drafts = total_drafts_per_day.sum()

    # Calculate the fill rate as the proportion of total drafts per day to total drafts
    fill_rate_per_day = total_drafts_per_day / total_drafts

    # Group by date and username to count unique draft_entry_id per user per day
    drafts_per_user_per_day = collective_df.groupby(['draft_date', 'username'])['draft_entry_id'].nunique().reset_index(name='drafts')

    # Calculate average drafts per user per day
    average_drafts_per_user_per_day = drafts_per_user_per_day.groupby('draft_date')['drafts'].mean()

    # Plotting
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # First axis for average drafts per user per day
    color = 'tab:red'
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Average Drafts per User Per Day', color=color)
    ax1.plot(average_drafts_per_user_per_day.index, average_drafts_per_user_per_day, color=color, marker='o', label='Avg Drafts/User/Day')
    ax1.tick_params(axis='y', labelcolor=color)

    # Second axis for the fill rate
    ax2 = ax1.twinx()
    color = 'tab:blue'
    ax2.set_ylabel('Fill Rate', color=color)
    ax2.plot(fill_rate_per_day.index, fill_rate_per_day, color=color, marker='o', label='Fill Rate')
    ax2.tick_params(axis='y', labelcolor=color)

    # Title and legend
    plt.title('Average Drafts Per User Per Day vs Fill Rate')
    fig.tight_layout()
    fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

    plt.show()

# Call the function with the updated DataFrame
plot_average_drafts_per_user_vs_fill_rate(collective_df)
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Convert draft_time to datetime and extract the draft_date
collective_df['draft_time'] = pd.to_datetime(collective_df['draft_time'], errors='coerce')
collective_df['draft_date'] = collective_df['draft_time'].dt.date

# Merge the DataFrames correctly
full_df = collective_df.merge(player_nfldraft_details_df, on=['player_name', 'position_name'], how='left')

# Exclude 'DJ Turner' from the dataset and prepare the 'is_rookie' column
full_df = full_df[full_df['player_name'] != 'DJ Turner']
full_df['is_rookie'] = full_df['Pick'].notnull()  # Set 'is_rookie' based on the presence of 'Pick'

# Check if the 'Pick' column is correctly merged
print("Columns in full_df after merge:", full_df.columns)

# Filter for drafts involving rookies and compute unique drafters
rookie_drafts_df = full_df[full_df['is_rookie']]
all_drafters_by_day = full_df.groupby('draft_date')['user_id'].nunique()

# Initialize dictionaries and sets for tracking
seen_rookie_drafters = set()
new_rookie_drafters_by_day = {}

# Process data to find new rookie drafters per day
for date, group in rookie_drafts_df.groupby('draft_date'):
    current_day_drafters = set(group['user_id'])
    new_drafters = current_day_drafters - seen_rookie_drafters
    new_rookie_drafters_by_day[date] = len(new_drafters)
    seen_rookie_drafters.update(new_drafters)

new_rookie_drafters_series = pd.Series(new_rookie_drafters_by_day).sort_index()

# Initialize the plot
fig = make_subplots(specs=[[{"secondary_y": True}]])
positions = ['RB', 'WR']
pick_bands = {'Early': (1, 63), 'Middle': (64, 102), 'Late': (103, float('inf'))}

# Add traces for ADP and rookie drafters
for position in positions:
    for band_name, (start_pick, end_pick) in pick_bands.items():
        band_df = full_df[(full_df['position_name'] == position) & (full_df['Pick'] >= start_pick) & (full_df['Pick'] <= end_pick)]
        adp_by_day = band_df.groupby('draft_date')['projection_adp'].mean().dropna()

        # Add traces for ADP
        fig.add_trace(
            go.Scatter(x=adp_by_day.index, y=adp_by_day, mode='lines+markers', name=f'{position} {band_name} Picks'),
            secondary_y=False
        )

# Add traces for new rookie drafters
fig.add_trace(
    go.Scatter(
        x=new_rookie_drafters_series.index,
        y=new_rookie_drafters_series,
        mode='lines',
        name='New Rookie Drafters',
        fill='tozeroy'
    ),
    secondary_y=True
)

# Update layout with titles and axis labels as before
fig.update_layout(
    title='ADP Over Time and New Unique Rookie Drafters Each Day',
    xaxis_title='Draft Date',
    yaxis_title='Average ADP',
    title_x=0.5,
    xaxis=dict(tickmode='auto', nticks=20, tickformat='%Y-%m-%d')
)

# Set secondary Y axis title for the new unique rookie drafters
fig.update_yaxes(title_text="Rookie Drafters Each Day", secondary_y=True)

# Correctly adding vertical lines and annotations
events = [
    {'date': '2023-04-27', 'label': 'Draft', 'buffer_space':150, 'xshift': 25},  # Example xshift
    {'date': '2023-05-08', 'label': 'Rookie Mini Camps Start', 'buffer_space':150, 'xshift': 85},  # Adjusted date for label position  # Increased xshift to move to the right
    {'date': '2023-06-18', 'label': 'Rookie Training Camps Start', 'buffer_space':150, 'xshift': 95},  # Increased xshift to move to the right
    {'date': '2023-08-29', 'label': 'Final Roster Cutdown', 'buffer_space':150, 'xshift': 70},
    # Add more events as needed
]
for event in events:
    fig.add_vline(x=pd.to_datetime(event['date']), line_width=2, line_dash="dash", line_color='red')
    fig.add_annotation(x=pd.to_datetime(event['date']), y=event['buffer_space'], text=event['label'],yshift=event['buffer_space'], xshift=event['xshift'], showarrow=False)

fig.show()
```
